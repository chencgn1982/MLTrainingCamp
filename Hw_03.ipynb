## 讀取資料

```python
"""讀取 train.csv, test.csv 並合併 """
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
df = df_train.append(df_test)
```

## 分離目標

```python
# 將目標分析欄位取出來 Get target columns
target = ['LotFrontage', 'GarageYrBlt', 'MasVnrArea']
df = df[target]
```

分別以平均值及中位數來填補缺失值

```python
df_NaMean = df.fillna(np.mean(df))
df_NaMedian = df.fillna(df.median(skipna=1))
```

## 繪圖

```python
fig = plt.figure(figsize=(10,10)) 
plt.subplot(331)
sns.histplot(df['LotFrontage'])
plt.subplot(332)
sns.histplot(df_NaMean['LotFrontage'])
plt.subplot(333)
sns.histplot(df_NaMedian['LotFrontage'])
plt.subplot(334)
sns.histplot(df['GarageYrBlt'])
plt.subplot(335)
sns.histplot(df_NaMean['GarageYrBlt'])
plt.subplot(336)
sns.histplot(df_NaMedian['GarageYrBlt'])
plt.subplot(337)
sns.histplot(df['MasVnrArea'])
plt.subplot(338)
sns.histplot(df_NaMean['MasVnrArea'])
plt.subplot(339)
sns.histplot(df_NaMedian['MasVnrArea'])
```

## 分析標準差

```python
for col in target:
    print(col+':')
    print('original: ',df[col].std())
    print('mean: ',df_NaMean[col].std())
    print('median: ',df_NaMedian[col].std())
    print('----------------------')
```
